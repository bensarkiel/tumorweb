```python
from tensorflow.keras.models import load_model

model = load_model('models/NASNet_2class_high.h5')

from collections import Counter

layer_types = [layer.__class__.__name__ for layer in model.layers]
counter = Counter(layer_types)

print("Jumlah layer per tipe:")
for layer_type, count in counter.items():
    print(f"{layer_type:<25}: {count}")

N = 10  # misalnya 10 layer terakhir
print(f"{'No.':<4} {'Layer Name':<30} {'Type':<25} {'Output Shape'}")
print("-" * 90)

for i, layer in enumerate(model.layers[-N:]):
    try:
        output_shape = layer.output.shape
    except AttributeError:
        output_shape = "N/A"
    print(f"{len(model.layers) - N + i:<4} {layer.name:<30} {layer.__class__.__name__:<25} {output_shape}")


important_layers = (
    "Conv2D", "SeparableConv2D", "Dense", "MaxPooling2D", 
    "AveragePooling2D", "GlobalAveragePooling2D", "Dropout", "BatchNormalization"
)

print(f"{'No.':<4} {'Layer Name':<30} {'Type':<25} {'Output Shape'}")
print("-" * 90)

for i, layer in enumerate(model.layers):
    layer_type = layer.__class__.__name__
    if layer_type in important_layers:
        try:
            output_shape = layer.output.shape
        except AttributeError:
            output_shape = "N/A"
        print(f"{i:<4} {layer.name:<30} {layer_type:<25} {output_shape}")

```
